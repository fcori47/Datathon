```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
hosp_train = pd.read_csv('Datasets/hospitalizaciones_train.csv')
hosp_train.isna().sum()
```

```python
hosp_train
```

```python
def transform_column_to_numbers(column):
    col_unique = column.unique()
    result = column.replace(col_unique,np.arange(len(col_unique)))
    return result
```

```python
ht = hosp_train.copy()
ht_str = ht.select_dtypes(include='object')
ht_str = ht_str.apply(lambda x: transform_column_to_numbers(x))
hosp_train['Stay (in days)'][hosp_train['Stay (in days)']<= 8]= 0
hosp_train['Stay (in days)'][hosp_train['Stay (in days)']> 8]= 1
for i in ht.columns:
    if i in ht_str.columns:
        hosp_train[i] = ht_str[i]

hosp_train
```

```python
decision = DecisionTreeClassifier(max_depth = 3, random_state = 42)

x = hosp_train[['Available Extra Rooms in Hospital', 'Department', 'Ward_Facility_Code',
       'doctor_name', 'staff_available', 'patientid', 'Age', 'gender',
       'Type of Admission', 'Severity of Illness', 'health_conditions',
       'Visitors with Patient', 'Insurance', 'Admission_Deposit']]

y= hosp_train['Stay (in days)']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=42)
```

```python
decision.fit(x_train.values,y_train.values)
```

```python
x_test = hosp_test[['Available Extra Rooms in Hospital', 'Department', 'Ward_Facility_Code',
       'doctor_name', 'staff_available', 'patientid', 'Age', 'gender',
       'Type of Admission', 'Severity of Illness', 'health_conditions',
       'Visitors with Patient', 'Insurance', 'Admission_Deposit']]
```

```python
print(decision.feature_importances_)
```

```python
y_pred = decision.predict(x_test.values)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_train[:90000],y_pred)
```
